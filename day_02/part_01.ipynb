```python
with open("inputs.txt","r") as fp:
    data = fp.read()
```

```python
player_map = {
    "X": "ROCK",
    "Y": "PAPER",
    "Z": "SCISSOR",
}
op_map = {
    "A": "ROCK",
    "B": "PAPER",
    "C": "SCISSOR",
}
score_map = {
    "ROCK": 1,
    "PAPER": 2,
    "SCISSOR": 3,
}
```

```python
score = 0
for line in data.strip().split("\n"):
    opponent, player = line.split(" ")
    opponent, player = op_map[opponent], player_map[player]
    
    score += score_map[player]

    if opponent == player:
        score += 3
    else:
        if (player == "ROCK" and opponent == "SCISSOR") or \
            (player == "PAPER" and opponent == "ROCK") or \
            (player == "SCISSOR" and opponent == "PAPER"):
            score += 6
            print(opponent, player, "WIN")
        else:
            print(opponent, player, "LOSE")
```
